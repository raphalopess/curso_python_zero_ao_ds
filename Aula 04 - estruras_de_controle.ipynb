```python
# Lógica E (AND)
# Lógica de multiplicação

# TRUE & TRUE
df_condicao_tt = df[( df['bedrooms'] == 4)  &  ( df['floors'] == 2 )] 

# TRUE & FALSE
df_condicao_tf = df[( df['bedrooms'] == 4)  &  ( df['floors'] == 20 )] 

# FALSE & FALSE
# Vai dar erro
df_condicao_ff = df[( df['bedrooms'] == 50)  &  ( df['floors'] == 20 )] 


df_condicao_tt.head()
```

```python
# Lógica Ou (OR)
# Lógica de soma

# TRUE & TRUE
df_condicao_tt = df[( df['bedrooms'] == 4)  |  ( df['floors'] == 2 )] 

# TRUE & FALSE
df_condicao_tf = df[( df['bedrooms'] == 4)  |  ( df['floors'] == 20 )] 

# FALSE & FALSE
# Vai dar erro
df_condicao_ff = df[( df['bedrooms'] == 50) |   ( df['floors'] == 20 )] 


df_condicao_tt.head()
```
