```python
# 1.	 Crise uma nova coluna chamada: "dormitory_type"
#     - Se o valor da cooluna "bedrooms" for igual à 1 => 'studio'
#     - Se o valor da cooluna "bedrooms" for igual à 2 => 'apartament'
#     - Se o valor da cooluna "bedrooms" for maior que 2 => 'house'

import pandas as pd

df = pd.read_csv(r"C:\Users\User\datasets\kc_house_data.csv")

df['date'] = pd.to_datetime(df ['date'])

df['price'] = df['price'].astype(float)

df['dormitory_type'] = 'standard'

# df.loc[df['price'] > 540000, 'level'] = 'high_level'

df.loc[df['bedrooms'] == 1, 'dormitory_type'] = 'studio'

df.loc[df['bedrooms'] == 2, 'dormitory_type'] = 'apartament'

df.loc[df['bedrooms'] > 2, 'dormitory_type']  = 'house'

df.head()
```

```python
# 2.	 Faça um gráfico de barras que representa a soma dos preços pelo numero de quartos.

# Gráfico de barras -> Modelo 
# df['year'] = pd.to_datetime(df['date']).dt.year
# by_year = df[['price','year']].groupby('year').sum().reset_index()
# plt.bar(by_year['year'], by_year['price'])

# importando uma biblioteca que possibilita criar gráficos
from matplotlib import pyplot as plt

by_bedrooms = df[['price','bedrooms']].groupby('bedrooms').sum().reset_index()
plt.bar(by_bedrooms['bedrooms'], by_bedrooms['price'])

plt.figure(figsize=(35,20))
```
